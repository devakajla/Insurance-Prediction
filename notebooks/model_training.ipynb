```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('features_insurance.csv')
```

```python
df.columns
```

```python
x=df[['age', 'bmi', 'children', 'smoke_enc', 'sex_enc', 'is_obese',
       'smoker_obese', 'northwest', 'southeast', 'southwest', 'bmi_smoker']]
y=df['charges']
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
r2=r2_score(y_test,y_pred_lr)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
rmse=root_mean_squared_error(y_test,y_pred_lr)
```

```python
rmse
```

```python
mae=mean_absolute_error(y_test,y_pred_lr)
print(f"MAE: {mae:.2f}")
```

```python
print(f"R2: {r2} \n, MAE : {mae} \n, RMSE: {rmse}")
```

## Ridge

```python
from sklearn.linear_model import Ridge

rg=Ridge()
```

```python
rg.fit(x_train,y_train)
```

```python
y_pred_rg=rg.predict(x_test)
```

```python
print(f"R2: {r2_score(y_test,y_pred_rg)}\n MAE: {mean_absolute_error(y_test,y_pred_rg)} \n RMSE: {root_mean_squared_error(y_test,y_pred_rg)}")
```

## Lasso

```python
from sklearn.linear_model import Lasso
```

```python
ls=Lasso(alpha=1.0)
```

```python
ls.fit(x_train,y_train)
```

```python
y_pred_ls=ls.predict(x_test)
```

```python
print(f"R2: {r2_score(y_test,y_pred_ls)}\n MAE: {mean_absolute_error(y_test,y_pred_ls)} \n RMSE: {root_mean_squared_error(y_test,y_pred_ls)}")
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(random_state=42)
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred_dt=dt.predict(x_test)
```

```python
print(f"R2: {r2_score(y_test,y_pred_dt)}\n MAE: {mean_absolute_error(y_test,y_pred_dt)} \n RMSE: {root_mean_squared_error(y_test,y_pred_dt)}")
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(n_estimators=200, max_depth=15, random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf=rf.predict(x_test)
```

```python
print(f"R2: {r2_score(y_test,y_pred_rf)}\n MAE: {mean_absolute_error(y_test,y_pred_rf)} \n RMSE: {root_mean_squared_error(y_test,y_pred_rf)}")
```

## XGBoost

```python
import xgboost as xgb
```

```python
xgb_model=xgb.XGBRegressor(n_estimators=200, max_depth=5, random_state=42)
```

```python
xgb_model.fit(x_train,y_train)
```

```python
y_pred_xg=xgb_model.predict(x_test)
```

```python
print(f"R2: {r2_score(y_test,y_pred_xg)}\n MAE: {mean_absolute_error(y_test,y_pred_xg)} \n RMSE: {root_mean_squared_error(y_test,y_pred_xg)}")
```
